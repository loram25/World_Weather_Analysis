```python
import pandas as pd
import gmaps
import requests
from citipy import citipy
from config import g_key
import numpy
from datetime import datetime
import time
from scipy.stats import linregress
```

```python
city_data=pd.read_csv('WeatherPy_vacation.csv', index_col=0)
city_data.head()
```

```python
au_cities=city_data.loc[(city_data['Country']=='AU')]
au_cities
```

```python
au_vac_cities=au_cities.loc[au_cities['City'].isin(['Nambucca Heads','Alice Springs','Ballina','Port Pirie'])]
au_vac_cities
```

```python
nambucca_heads=au_cities.loc[au_cities['City']=='Nambucca Heads']
alice_springs=au_cities.loc[au_cities['City']=='Alice Springs']
ballina=au_cities.loc[au_cities['City']=='Ballina']
port_pirie=au_cities.loc[au_cities['City']=='Port Pirie']

ballina
```

```python
import gmaps.datasets
gmaps.configure(api_key=g_key)
```

```python
nambucca_heads_t=(-30.65,153.00)
alice_springs_t=(-23.70,133.88)
ballina_t=(-28.87,153.57)
port_pirie_t=(-33.18,138.02)

fig=gmaps.figure()
au_map=gmaps.directions_layer(nambucca_heads_t, alice_springs_t, waypoints=[ballina_t,port_pirie_t], travel_mode='DRIVING')
fig.add_layer(au_map)
fig
```

```python
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Current Weather</dt><dd>{Current Description} and {Max Temp} °F</dd>
</dl>
"""
hotel_info=[info_box_template.format(**row) for index, row in au_vac_cities.iterrows()]
```

```python
locations=au_vac_cities[['Lat','Lng']]
fig=gmaps.figure(center=(30.0,31.0), zoom_level=1.5)
marker_layer=gmaps.marker_layer(locations, info_box_content=hotel_info)
fig.add_layer(marker_layer)
fig
```
